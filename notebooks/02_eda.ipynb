```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('C:\\Users\\DVK\\Desktop\\EMIPredict AI\\data\\emi_prediction_dataset_cleaned.csv')
df.head()
```

```python
df['emi_eligibility'].value_counts()
sns.countplot(x='emi_eligibility', data=df)
plt.title("EMI Eligibility Distribution")
plt.show()
```

```python
df['emi_scenario'].value_counts()
sns.countplot(y='emi_scenario', data=df)
plt.title("EMI Scenario Distribution")
plt.show()
```

```python
sns.histplot(df['age'], bins=20, kde=True)
plt.title("Age Distribution of Customers")
plt.show()
```

```python
sns.histplot(df['monthly_salary'], bins=30, kde=True)
plt.title("Monthly Salary Distribution")
plt.show()
```

```python
sns.boxplot(x=df['monthly_salary'])
plt.title("Monthly Salary Outliers")
plt.show()
```

```python
sns.boxplot(x='emi_eligibility', y='monthly_salary', data=df)
plt.title("Salary vs EMI Eligibility")
plt.show()
```

```python
sns.histplot(df['credit_score'], bins=30, kde=True)
plt.title("Credit Score Distribution")
plt.show()
sns.boxplot(x='emi_eligibility', y='credit_score', data=df)
plt.title("Credit Score vs EMI Eligibility")
plt.show()
```

```python
df['total_monthly_expense'] = (
    df['school_fees'] +
    df['college_fees'] +
    df['travel_expenses'] +
    df['groceries_utilities'] +
    df['other_monthly_expenses']
)
sns.histplot(df['total_monthly_expense'], bins=30, kde=True)
plt.title("Total Monthly Expense Distribution")
plt.show()
```

```python
sns.scatterplot(
    x='monthly_salary',
    y='total_monthly_expense',
    hue='emi_eligibility',
    data=df,
    alpha=0.5
)
plt.title("Income vs Expense by EMI Eligibility")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(
    df.select_dtypes(include=['int64','float64']).corr(),
    cmap='coolwarm',
    annot=False
)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.histplot(df['max_monthly_emi'], bins=30, kde=True)
plt.title("Max Monthly EMI Distribution")
plt.show()
sns.scatterplot(
    x='monthly_salary',
    y='max_monthly_emi',
    data=df
)
plt.title("Salary vs Max EMI")
plt.show()
```
